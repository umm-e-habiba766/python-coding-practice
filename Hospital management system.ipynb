```python
class Hospital:
    def __init__(self, hospital_name):
        self.hospital_name = hospital_name
        self.patients = []
        self.doctors = []

    def add_patient(self, name, age, disease):
        patient = {"name": name, "age": age, "disease": disease}
        self.patients.append(patient)

    def add_doctor(self, name, specialty):
        doctor = {"name": name, "specialty": specialty}
        self.doctors.append(doctor)

    def check_patient_condition(self, age, critical, appointment):
        if critical.lower() == "yes":
            critical = True 
        else:
            critical = False
        if appointment.lower() == "yes":
            appointment = True
        else:
            appointment = False

        if age > 60 or critical:
            print("Immediate Treatment")
        elif appointment:
            print("Treated after Priority Cases")
        else:
            print("General Queue")

    def show_patients(self):
        if not self.patients:
            print("No patients registered yet!")
        else:
            print("\n--- Patient List ---")
            for p in self.patients:
                print(f"Name: {p['name']}, Age: {p['age']}, Disease: {p['disease']}")

    def show_doctors(self):
        if not self.doctors:
            print("No doctors registered yet!")
        else:
            print("\n--- Doctor List ---")
            for d in self.doctors:
                print(f"Name: {d['name']}, Specialty: {d['specialty']}")

    def search_patient(self, name):
        found = False
        for p in self.patients:
            if p['name'].lower() == name.lower():  # case-insensitive search
                print(f"Patient Found → Name: {p['name']}, Age: {p['age']}, Disease: {p['disease']}")
                found = True
                break
        if not found:
            print("Patient not found.")

    def search_doctor(self, name):
        found = False
        for d in self.doctors:
            if d['name'].lower() == name.lower():  # case-insensitive search
                print(f"Doctor Found → Name: {d['name']}, Specialty: {d['specialty']}")
                found = True
                break
        if not found:
            print("Doctor not found.")



hospital = Hospital("City Hospital")

while True:
    print(f"\n--- {hospital.hospital_name} Management System ---")
    print("1. Add Patient")
    print("2. Add Doctor")
    print("3. Show All Patients")
    print("4. Show All Doctors")
    print("5. Search Patient")
    print("6. Search Doctor")
    print("7. Check Patient Condition")
    print("8. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        disease = input("Enter patient disease: ")
        hospital.add_patient(name, age, disease)
        print("Patient added successfully.")

    elif choice == "2":
        name = input("Enter doctor name: ")
        specialty = input("Enter doctor specialty: ")
        hospital.add_doctor(name, specialty)
        print("Doctor added successfully.")

    elif choice == "3":
        hospital.show_patients()

    elif choice == "4":
        hospital.show_doctors()

    elif choice == "5":
        name = input("Enter patient name to search: ")
        hospital.search_patient(name)

    elif choice == "6":
        name = input("Enter doctor name to search: ")
        hospital.search_doctor(name)

    elif choice == "7":
        age = int(input("Enter patient age: "))
        critical = input("Is patient critical? (yes/no): ")
        appointment = input("Does patient have appointment? (yes/no): ")
        hospital.check_patient_condition(age, critical, appointment)

    elif choice == "8":
        print("Exiting Hospital System...")
        break

    else:
        print("Invalid choice! Try again.")
```
